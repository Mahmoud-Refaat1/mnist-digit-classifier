```python
import numpy as np         
import pandas as pd
import matplotlib.pyplot as plt      
import random          
```

```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
plt.imshow(train_images[0], cmap='gray')
plt.title(f'Label: {train_labels[0]}')
plt.show()
```

```python
train_images = train_images.reshape((60000, 784)).astype('float32') / 255
test_images = test_images.reshape((10000, 784)).astype('float32') / 255


train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test accuracy:', test_acc)
```

```python
# Choose any img number to classify
index = 0  # From 0 to 9999
image = test_images[index]
label = np.argmax(test_labels[index])  # التصنيف الحقيقي

# reshape the img to (1, 784)
image_reshaped = image.reshape(1, 784)

# Use Model To Predict
prediction = model.predict(image_reshaped)
predicted_label = np.argmax(prediction)

# Show Img
plt.imshow(image.reshape(28, 28), cmap='gray')
plt.title(f"Actual: {label} | Predicted: {predicted_label}")
plt.show()
```

```python
#Show First 10 imgs
for i in range(10):
    image = test_images[i].reshape(1, 784)
    prediction = model.predict(image)
    predicted_label = np.argmax(prediction)
    true_label = np.argmax(test_labels[i])

    plt.imshow(test_images[i].reshape(28, 28), cmap='gray')
    plt.title(f"Actual: {true_label} | Predicted: {predicted_label}")
    plt.axis('off')
    plt.show()
```
